# 正規分布

## 1次元について

$$
p(x) = \frac{1}{\sqrt{2\pi\sigma^{2}}}\exp\left(-\frac{(x - \mu)^{2}}{2\sigma^{2}}\right)
$$

### この関数の解析

#### ステップ1 ただの指数関数

$$
f(x) = \exp\left(x\right)
$$

```python
import numpy as np
import matplotlib.pyplot as plt

f = lambda x: np.exp(x)
xs = np.arange(0., 10., 0.2)
fxs = np.array([f(x) for x in xs])
plt.plot(xs, fxs)
plt.show()
```

#### ステップ2 xについて負にする

$$
f(x) = \exp\left(-x\right)
$$

```python
f = lambda x: np.exp(-x)
xs = np.arange(-10., 10., 0.2)
fxs = np.array([f(x) for x in xs])
plt.plot(xs, fxs)
plt.show()
```

#### ステップ3 2乗する

$$
f(x) = \exp\left(-x^{2}\right)
$$

```python
f = lambda x: np.exp(-x**2)
xs = np.arange(-10., 10., 0.2)
fxs = np.array([f(x) for x in xs])
plt.plot(xs, fxs)
plt.show()
```

#### ステップ4 平行移動を表現する

$$
f(x) = \exp\left(-(x - \mu)^{2}\right)
$$

```python
f = lambda x, u: np.exp(-(x - u)**2)
xs = np.arange(-10., 10., 0.2)
fxs = np.array([f(x, 3.) for x in xs])
plt.plot(xs, fxs)
plt.show()
```

#### ステップ4 幅を変更できるようにする

$$
f(x) = \exp\left(-\frac{(x - \mu)^{2}}{2\sigma^{2}}\right)
$$

```python
f = lambda x, u, s: np.exp(-((x - u)**2)/(2. * s**2.))
xs = np.arange(-10., 10., 0.2)
fxs = np.array([f(x, 3., 3.) for x in xs])
plt.plot(xs, fxs)
plt.show()
```

#### ステップ5 積分して1になるように正規化

$$
p(x) = \frac{1}{\sqrt{2\pi\sigma^{2}}}\exp\left(-\frac{(x - \mu)^{2}}{2\sigma^{2}}\right)
$$

```python
import numpy as np
import matplotlib.pyplot as plt

u = 3.
s = 2.
p = lambda x, u, s: 1. / np.sqrt(2. * np.pi * s**2) * np.exp(-((x - u)**2) / (2. * s**2))

xs = np.arange(-10., 15., 0.2)
pxs = np.array([p(x, u, s) for x in xs])
plt.plot(xs, pxs)
plt.show()
```

設定した正規分布に従う確率密度関数を使ってデータを収集してみる

```python
pxs = np.array([np.random.normal(u, s) for i in range(100)])
plt.plot(pxs)
plt.show()
```

\begin{eqnarray}

    \sigma^{2} & = & \frac{1}{N-1}\sum^{N-1}_{n=1}(x - \mu)^{2}
\end{eqnarray}
